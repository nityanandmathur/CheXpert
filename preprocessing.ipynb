NaNs and -1s to ZEROS

```python
def fill(df):
    return df.fillna(0)
```

```python
train_df = fill(train_df)
val_df = fill(val_df)
test_df = fill(test_df)
```

```python
train_df
```

```python
def zeroing(df):
    return df.replace(-1, 0)
```

```python
train_df = zeroing(train_df)
val_df = zeroing(val_df)
test_df = zeroing(test_df)
```

```python
train_df
```

```python
reordered_columns = [
    "Path",
    "Atelectasis",
    "Cardiomegaly",
    "Consolidation",
    "Edema",
    "Pleural Effusion",
    "Pleural Other",
    "Pneumonia",
    "Pneumothorax",
    "Enlarged Cardiomediastinum",
    "Lung Opacity",
    "Lung Lesion",
    "Fracture",
    "Support Devices",
    "No Finding"
]
```

```python
def reorder(df):
    return df[reordered_columns]
```

```python
train_df = reorder(train_df)
val_df = reorder(val_df)
test_df = reorder(test_df)
```

```python
train_df
```

```python
def print_count(df):
    for i in range(1,len(df.columns)):
        print(df.columns[i], df[df.columns[i]].sum())
```
